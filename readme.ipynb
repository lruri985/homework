## <img src="https://raw.githubusercontent.com/FortAwesome/Font-Awesome/6.x/svgs/solid/database.svg" width="20" height="20"> Dataset

 - 這個資料集是 Meta 廣告投放的成效資訊，總共有 64,827 筆

 - 每筆資料表示一個廣告 (ad_id) 在某一段時間投放在某個廣告平臺 (publisher_platform) 中某個版位 (platform_position) 的點擊率 (ctr)
 
 - 這些廣告被投放的時間點分佈在 2023 年左右

 - 每一個廣告投放的時間長度不一樣、投放的對象設置、語言、國家也不盡相同，這份資料並沒有提供這些資訊，而且很大程度與投放初期的 ctr 有關

 - 資料集提供的廣告素材資訊包含
 
   - 廣告素材原始圖檔 (image_hash)
  
   - 廣告預覽截圖 (preview_hash)
  
   - 廣告標題 (title)
  
   - 廣告內文 (body)

 - 有些廣告會反覆使用相同的素材，也就是同樣的 image、preview、title、body 等可能被運用在不同的廣告中

### How to load this data

```python
import pandas as pd

insights = pd.read_pickle("insights.pkl")
images = pd.read_pickle("images.gzip", compression="gzip")
previews = pd.read_pickle("previews.gzip", compression="gzip")

dat = pd.merge(insights, images, on="image_hash", how="left")
dat = pd.merge(dat, previews, on="preview_hash", how="left")
del insights, images, previews
```

### Schema

#### insight 資料集

|Column|Description|
|------|-----------|
|ad_id|共 12,104 個廣告編號，每一個廣告編號可能使用相同的素材，也就是 image、preview、title、body 可能重複使用在不同的廣告 ad_id 中。|
|image_hash|廣告圖片編號，代表廣告的原始素材，可能為空值，並且與 image dataset 中的同名欄位對應。|
|preview_hash|廣告預覽截圖編號，代表用戶實際在行動裝置上所看到的畫面，無空值，並且與 preview dataset 中的同名欄位對應。|
|publisher_platform|共 5 個廣告平臺，分為：facebook (54%)、instagram (41%)、audience_network (3.4%)、messenger (1.2%)、unknown (0.14%) 。|
|platform_position|共 27 個廣告平臺中的版位，facebook 包含：feed、marketplace、video_feeds 等，、instagram 包含：feed、instagram_stories、instagram_explore 等。|
|ctr|表示該廣告被投放在廣告平臺中的版位後被點擊的比例，通常在 0.005 左右表示為 0.5%，點擊率的計算是：點擊次數 / 曝光次數，這個資料集中沒有提供點擊次數、與曝光次數。|
|reach|表示該廣告投放後被多少 Meta 帳號看過，同一個帳號可以反覆看到重複的廣告，所以不代表曝光次數，受到廣告預算、投放時間與設置影響，且每一個廣告都至少觸及超過 10 個 meta 帳號。|
|title|899 個廣告標題，包含不同種語言。|
|body|1,730 個廣告內文，包含不同種語言；廣告標題X廣告內文 共有 1,924 個組合。|

#### image 資料集

|Column|Description|
|------|-----------|
|image_hash|廣告圖片編號，代表廣告的原始素材，並且與 insight dataset 中的同名欄位對應。|
|image|廣告圖片，為 PIL.JpegImagePlugin.JpegImageFile 型別。|

#### preview 資料集

|Column|Description|
|------|-----------|
|preview_hash|廣告預覽截圖編號，代表用戶實際在行動裝置上所看到的畫面，並且與 insight dataset 中的同名欄位對應。|
|preview|廣告預覽截圖編號，為 PIL.JpegImagePlugin.JpegImageFile 型別。|
